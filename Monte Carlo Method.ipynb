```python
%matplotlib inline
import matplotlib.pyplot as plt
import random
import math
import numpy as np

sample_num = 10000 # Number of darts thrown in the simulation
hit_xcoordinates = []
hit_ycoordinates = []
miss_xcoordinates = []
miss_ycoordinates = []
    
for i in range(sample_num):
    x_coordinate = random.random()
    y_coordinate = random.random()
    hit = math.hypot(x_coordinate, y_coordinate) <= 1
    if hit:
        hit_xcoordinates.append(x_coordinate)
        hit_ycoordinates.append(y_coordinate)
    else:
        miss_xcoordinates.append(x_coordinate)
        miss_ycoordinates.append(y_coordinate)

plt.rcParams["figure.figsize"] = (8,8)
plt.plot(hit_xcoordinates, hit_ycoordinates, 'o')
plt.plot(miss_xcoordinates, miss_ycoordinates, 'o')
plt.title( \
'Figure 5. Visualization of a revised simulation (N = 10,000)')
plt.show()
```
